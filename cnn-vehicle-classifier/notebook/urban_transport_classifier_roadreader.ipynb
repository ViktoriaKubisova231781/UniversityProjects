````python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
import time

def download_image(image_url, directory, index):
    """
    Downloads an image from a given URL into a specified directory.
    """
    try:
        response = requests.get(image_url)
        if response.status_code == 200:
            file_path = os.path.join(directory, f'image_{index}.jpg')
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f"Downloaded {file_path}")
    except Exception as e:
        print(f"Error downloading {image_url}: {e}")

def scrape_images(search_term, limit):
    """
    Scrapes images from DuckDuckGo based on a search term up to a specified limit.
    """
    # Initialize the Chrome driver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    # Go to DuckDuckGo's image search page
    driver.get(f"https://duckduckgo.com/?q={search_term}&iax=images&ia=images")
    time.sleep(3)  # Allow time for the page to load

    # Scroll the page to ensure that images are loaded
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Wait for new images to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    # Find image elements on the page
    images = driver.find_elements(By.CSS_SELECTOR, "img.tile--img__img.js-lazyload")
    images = images[:limit]  # Limit the number of images to download

    # Create a directory for the downloaded images
    directory = f"downloaded_images_{search_term.replace(' ', '_')}"
    if not os.path.exists(directory):
        os.makedirs(directory)

    # Download the images
    for index, image in enumerate(images):
        image_url = image.get_attribute('src')
        if image_url:
            download_image(image_url, directory, index)

    driver.quit()
    print(f"Finished downloading images to {directory}.")

# Example usage
search_term = "car" # Specify the search term here
limit = 100  # Specify the maximum number of images to download

scrape_images(search_term, limit)
````

Given the absence of bias in my dataset, I decided to extend my analysis to the full Imsitu dataset, aiming to identify and analyze potential instances of bias present therein.

One of the biases observed in the Imsitu dataset emerged during a search for images associated with the verb "cooking." The dataset contained 81 images of women cooking, contrasting with only 37 images of men engaged in the same activity. This bias can be categorized as **historical or pre-existing bias** tied to societal gender perceptions and stereotypes surrounding cooking. Traditionally, cultural norms have linked cooking more closely with women, femininity, or domesticity, reflecting in the overrepresentation of women in the dataset. Throughout history, women have often been associated with domestic roles involving cooking, whereas men have been less prominently depicted in cooking-related activities.

A similar bias was evident when searching for the verb "dusting," with 120 images of women dusting and only 26 images of men engaged in the same activity. Conversely, when searching for verbs associated with activities culturally associated with men, such as "constructing", "repairing", or "loading", there tended to be approximately twice as many images of men as women, reflecting pre-existing gender biases in those specific contexts.

This bias is undesirable as it perpetuates gender stereotypes, with women overrepresented in domestic chores like cooking and dusting, while men are depicted more often in manual labor or technical activities. This perpetuation reinforces traditional gender roles and societal expectations, limiting individuals' freedom and contributing to gender inequalities. Moreover, biased datasets can lead to unfair outcomes in machine learning applications, reinforcing discriminatory practices in areas like employment and education. Ultimately, this bias disproportionately affects individuals by reinforcing stereotypes, limiting opportunities, and potentially perpetuating discrimination and inequality, particularly in decision-making processes and system development.

Another type of bias identified within the Imsitu dataset was the imbalance in the representation of different racial groups. When searching for pictures of a man, out of the 24,894 found images, the majority depicted white men, with only a few labeled as "man" were representing black men. Similarly, when searching for pictures of a "woman", out of the 18,289 images found, only a few were of black women. Specifically searching for "black man" yielded only 25 images, and searching for "black woman" resulted in only 22 images.

This racial bias can be categorized under **representation bias**, where certain groups are underrepresented or overrepresented compared to their actual prevalence in the population. In this case, the overrepresentation of white people and the underrepresentation of black people in the dataset suggest a bias that may stem from various factors, including the demographics of the image sources, selection criteria for including images, or societal biases and stereotypes about race. In this case, the imbalance in racial representation  skews the dataset's composition and may lead to biased model outcomes.

The overrepresentation of white individuals and the underrepresentation of black individuals is undesirable for several reasons. Firstly, it perpetuates and reinforces societal stereotypes and biases regarding race, as it suggests that white people are more commonly associated with certain activities depicted in the dataset. Once we've gathered and explored the data for our machine learning model, it's essential to establish baselines to lay the foundation for our deep learning model.

Baselines are pivotal in any machine learning endeavor for several reasons. Firstly, they serve as a benchmark against which we can compare the performance of more advanced models. Secondly, baselines assist in setting realistic performance expectations. They provide insight into what level of performance is easily achievable, which, in turn, informs decisions regarding resource allocation and project direction. Furthermore, if a complex model only marginally surpasses a simple baseline, it prompts a reassessment of the approach, potentially saving time and resources.

The Random Guess Baseline is a simple baseline model used in machine learning tasks for comparison purposes. It involves making predictions or classifications randomly, without any consideration for the data or problem at hand. Essentially, it offers a basic probability reference if the model made predictions completely at random, without any learning or pattern recognition from the data.

**In multi-class classification tasks with *n* balanced classes, the random guess baseline would predict each class with a probability of $\frac{1}{n}$.**

It serves as a benchmark to assess the performance of more sophisticated models. A model should perform significantly better than the random guess baseline to be considered effective.

````python
from skimage import io as ski_io
import os
import numpy as np
import skimage as ski

# Folder paths where the images of each class are stored
folder_paths = [
    r"/content/drive/MyDrive/ImageClassifierDataset/car",
    r"/content/drive/MyDrive/ImageClassifierDataset/bus",
    r"/content/drive/MyDrive/ImageClassifierDataset/bicycle",
    r"/content/drive/MyDrive/ImageClassifierDataset/train",
    r"/content/drive/MyDrive/ImageClassifierDataset/van",
    r"/content/drive/MyDrive/ImageClassifierDataset/truck",
    r"/content/drive/MyDrive/ImageClassifierDataset/motorcycle",
    r"/content/drive/MyDrive/ImageClassifierDataset/scooter"
]

# Class labels for the images (same order as folder_paths)
class_names = [
    "car",
    "bus",
    "bicycle",
    "train",
    "van",
    "truck",
    "motorcycle",
    "scooter"
]

# Target size to resize the images
target_size = (256, 256)

# Arrays to store the images (X) and class labels (y)
X = []
y = []

for i in range(len(folder_paths)):
    for filename in os.listdir(folder_paths[i]):
        file_extension = os.path.splitext(filename)[1]
        # Check if the file extension is an image extension
        if file_extension == ".jpg" or file_extension == ".jpeg" or file_extension == ".png":
            image_filename = os.path.join(folder_paths[i], filename)
            # Read the image file using scikit-image
            image = ski.io.imread(image_filename)
            # Resize the image to the target_size
            image_resized = ski.transform.resize(image, target_size)
            if image_resized.shape == (256, 256, 3):
                # Normalize the image
                image_normalized = image_resized / 255.0
                # Insert the image into array X
                X.append(image_normalized)
                # Insert the class label into array y
                y.append(class_names[i])
            else:
                print(filename)

# Convert of arrays into numpy arrays
X_MLP = np.array(X)
y_MLP = np.array(y)
````

````python
# Split the data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_MLP, y_MLP, test_size = 1/10, random_state = 0)
print(X_train.shape, X_test.shape)

# Split the data into train, test and validations sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size = 2/9, random_state = 0)
print("Shape of X_train before resizing:", X_train.shape)
print("Shape of X_test before resizing:", X_test.shape)
print("Shape of X_val before resizing:", X_val.shape)

# Reshape the input data
X_train_reshaped = X_train.reshape(-1,256*256*3)
X_test_reshaped = X_test.reshape(-1, 256*256*3)
X_val_reshaped = X_val.reshape(-1, 256*256*3)

# Print the shapes of reshaped input data
print("Shape of X_train after resizing:", X_train_reshaped.shape)
print("Shape of X_test after resizing:", X_test_reshaped.shape)
print("Shape of X_val after resizing:", X_val_reshaped.shape)

# Initialize LabelEncoder
label_encoder = preprocessing.LabelEncoder()

# Fit and transform integer labels for training set
y_train_integer_encoded = label_encoder.fit_transform(y_train)
y_train_onehot_encoded = to_categorical(y_train_integer_encoded, num_classes=8)
print("Shape of y_train_onehot_encoded:", y_train_onehot_encoded.shape)

# Fit and transform integer labels for test set
y_test_integer_encoded = label_encoder.transform(y_test)
y_test_onehot_encoded = to_categorical(y_test_integer_encoded, num_classes=8)
print("Shape of y_test_onehot_encoded:", y_test_onehot_encoded.shape)

# Fit and transform integer labels for validation set
y_val_integer_encoded = label_encoder.transform(y_val)
y_val_onehot_encoded = to_categorical(y_val_integer_encoded, num_classes=8)
print("Shape of y_val_onehot_encoded:", y_val_onehot_encoded.shape)
````

````python
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense

# Build the Baseline MLP model
model_MLP = Sequential([
    Dense(128, activation="relu", input_shape=(X_train_reshaped.shape[1], )),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(8, activation="softmax")
])

# Compile the model
model_MLP.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Train the model
history_MLP = model_MLP.fit(X_train_reshaped, y_train_onehot_encoded, epochs=30, batch_size=32,
                    validation_data=(X_val_reshaped, y_val_onehot_encoded))

# Evaluate the model
loss, accuracy = model_MLP.evaluate(X_test_reshaped, y_test_onehot_encoded)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
````

To implement early stopping, we'll train the model by fitting it to our training data while initializing the early_stopping callback from the outset.

````python
from keras.callbacks import EarlyStopping

# Build the Baseline MLP model
model_MLP = Sequential([
    Dense(128, activation="relu", input_shape=(X_train_reshaped.shape[1], )),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(8, activation="softmax")
])

# Define the early stopping callback
early_stopping = EarlyStopping(monitor='val_accuracy', patience=2, restore_best_weights=True)

# Compile the model
model_MLP.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model with early stopping
history_MLP = model_MLP.fit(X_train_reshaped, y_train_onehot_encoded, epochs=30, batch_size=32,
                    validation_data=(X_val_reshaped, y_val_onehot_encoded), callbacks=[early_stopping])

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot learning curves for accuracy and validation accuracy
axs[0].plot(history_MLP.history["accuracy"], label="Train Accuracy")
axs[0].plot(history_MLP.history["val_accuracy"], label="Validation Accuracy")
axs[0].set_xlabel("Epochs")
axs[0].set_ylabel("Accuracy")
axs[0].set_title("Accuracy vs. Epochs")
axs[0].legend()

# Plot learning curves for loss and validation loss
axs[1].plot(history_MLP.history["loss"], label="Train Loss")
axs[1].plot(history_MLP.history["val_loss"], label="Validation Loss")
axs[1].set_xlabel("Epochs")
axs[1].set_ylabel("Loss")
axs[1].set_title("Loss vs. Epochs")
axs[1].legend()

plt.tight_layout()
plt.show()

# Evaluate the model on the test set
loss, accuracy = model_MLP.evaluate(X_test_reshaped, y_test_onehot_encoded)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
````

In the Machine Learning Project Lifecycle, low model accuracy often results from insufficient data. To address this in the second iteration, we plan to use the same model but with a larger dataset. To expedite this process without the time-consuming task of gathering and labeling new data, we will employ data augmentation.

Data augmentation is a technique used to artificially increase the size of a dataset by generating modified versions of existing data. It is often used in machine learning when the original dataset is small to prevent overfitting and improve the model's generalization.

Various methods exist for augmenting data, tailored to the data type and task. For our dataset, we'll implement the following augmentation techniques:
- Horizontal image flipping
- Image rotation by a specified angle
- Image zooming by a specified factor

Keras provides several ways to apply common augmentation techniques. Because the `ImageDataGenerator` method is deprecated, we'll utilize Keras preprocessing layers for data augmentation during image preprocessing. Each image will be augmented once and added to the dataset alongside the original images. The remaining preprocessing steps, including resizing, one-hot encoding and data splitting, will remain unchanged from the first iteration.

````python
import os
import numpy as np
import skimage as ski
from skimage import transform
from sklearn import preprocessing
from keras.utils import to_categorical
from tensorflow import keras
from tensorflow.keras import layers

# Folder paths where the images of each class are stored
folder_paths = [
    r"/content/drive/MyDrive/ImageClassifierDataset/car",
    r"/content/drive/MyDrive/ImageClassifierDataset/bus",
    r"/content/drive/MyDrive/ImageClassifierDataset/bicycle",
    r"/content/drive/MyDrive/ImageClassifierDataset/train",
    r"/content/drive/MyDrive/ImageClassifierDataset/van",
    r"/content/drive/MyDrive/ImageClassifierDataset/truck",
    r"/content/drive/MyDrive/ImageClassifierDataset/motorcycle",
    r"/content/drive/MyDrive/ImageClassifierDataset/scooter"
]

# Class labels for the images (same order as folder_paths)
class_names = [
    "car",
    "bus",
    "bicycle",
    "train",
    "van",
    "truck",
    "motorcycle",
    "scooter"
]

# Target size to resize the images
target_size = (256, 256)

# Arrays to store the images (X) and class labels (y)
augumented_X = []
augumented_y = []

# Creates the preprocessing layer for data augmentation
data_augmentation = keras.Sequential(
    [
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.2),
    ])

for i in range(len(folder_paths)):
    for filename in os.listdir(folder_paths[i]):
        file_extension = os.path.splitext(filename)[1]
        # Check if the file extension is an image extension
        if file_extension == ".jpg" or file_extension == ".jpeg" or file_extension == ".png":
            image_filename = os.path.join(folder_paths[i], filename)
            # Read the image file using scikit-image
            image = ski.io.imread(image_filename)
            # Resize the image to the target_size
            image_resized = ski.transform.resize(image, target_size)
            if image_resized.shape == (256, 256, 3):
                # Append original image to X and corresponding label to y
                augumented_X.append(image_resized)
                augumented_y.append(class_names[i])
                # Apply augmentation once
                augmented_images = data_augmentation(np.expand_dims(image_resized, axis=0))
                # Insert the augmented images into array X
                for augmented_image in augmented_images:
                    augumented_X.append(augmented_image)
                    # Insert the class label into array y
                    augumented_y.append(class_names[i])
            else:
                print(filename)

# Convert arrays into numpy arrays
augumented_X = np.array(augumented_X)
augumented_y = np.array(augumented_y)

# Initialize LabelEncoder with class names
label_encoder = preprocessing.LabelEncoder()
label_encoder.classes_ = np.array(class_names)

# Encode class labels
augumented_y_integer_encoded = label_encoder.transform(augumented_y)

# Convert integer-encoded labels into one-hot encoded vectors
augumented_y_onehot_encoded = keras.utils.to_categorical(augumented_y_integer_encoded, num_classes=len(class_names))
print("Shape of augumented y_onehot_encoded:", augumented_y_onehot_encoded.shape)

# Split the data into train and test sets
from sklearn.model_selection import train_test_split
augumented_X_train, augumented_X_test, augumented_y_train, augumented_y_test = train_test_split(augumented_X, augumented_y_onehot_encoded, test_size=1/10, random_state=0)

# Split the data into train, test, and validation sets
augumented_X_train, augumented_X_val, augumented_y_train, augumented_y_val = train_test_split(augumented_X_train, augumented_y_train, test_size=2/9, random_state=0)
print("Shape of augumented X_train:", augumented_X_train.shape)
print("Shape of augumented X_val:", augumented_X_val.shape)
print("Shape of augumented X_test:", augumented_X_test.shape)
````

The model we used is a convolutional neural network (CNN) designed for image classification tasks. It follows a sequential architecture, which means the layers are stacked linearly.

The model starts with a `Conv2D` layer with 32 filters of size (3, 3), employing ReLU activation function. This layer processes the input images with a stride of (2, 2) and preserves the input shape through padding. Following the convolutional layer, a `MaxPooling2D` layer with a pool size of (2, 2) is applied to reduce spatial dimensions, effectively downsampling the feature maps. The subsequent layers repeat the pattern: another `Conv2D` layer with 64 filters, ReLU activation, and the same padding and stride settings, followed by another `MaxPooling2D` layer. After the convolutional and pooling layers, the feature maps are flattened into a 1D array using the `Flatten` layer. A dropout layer with a dropout rate of 0.5 is introduced to mitigate overfitting by randomly dropping 50% of the neurons during training. The flattened features are then passed through a fully connected `Dense` layer with 128 neurons and ReLU activation. Finally, the output layer consists of a `Dense` layer with 8 neurons, one for each class in the classification task, using the softmax activation function to output the probabilities of each class.

The model summary, generated using `model.summary()`, provides a concise overview of the architecture, including the layer types, output shapes, and the total number of parameters to be trained, which sums up to 2,117,704.

````python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dropout, Dense

# Instantiate the model
V1_model = Sequential([
    Conv2D(32, (3, 3), activation='relu',
           input_shape=(256, 256, 3),
           padding='same', strides=(2, 2)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu',
           padding='same', strides=(2, 2)),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dropout(0.5),
    Dense(128, activation='relu'),
    Dense(8, activation='softmax')
])

#  Summarize the model
V1_model.summary()
````

Additionally, we specify accuracy as the metric to monitor the model's performance during training, allowing us to track how well the model is classifying the images.

Next, we instantiate an EarlyStopping callback to prevent overfitting and efficiently halt training when the model's performance on the validation set stops improving. We monitor the validation loss, setting a patience of 2 epochs, which means training will stop if there is no improvement in validation loss for two consecutive epochs. Additionally, we set restore_best_weights=True to ensure that the model reverts to the weights yielding the lowest validation loss during training.

Finally, we train the model using the fit method, specifying 20 epochs and a batch size of 32. This configuration allows the model to iterate over the entire training dataset 20 times, adjusting the weights based on the calculated loss and optimizing the network's performance. We also provide the validation data to evaluate the model's performance on unseen data during training. The EarlyStopping callback is included in the training process to dynamically halt training when necessary, preventing overfitting and ensuring the model generalizes well to new data.

After loading and preprocessing the data, the next step involves defining the functions that constitute the Multilayer Perceptron (MLP) algorithm. Leveraging the functions constructed in the "MLP from Scratch" notebook, I made adjustments to accommodate my dataset, particularly in the loss function. This adjustment allows the model to handle multiple classes during training and evaluation.

````python
from copy import deepcopy

def initializer(nn_architecture):
    '''
    Given
    - the number of layers
    - the number of units in each layer
    initializes network parameters (weights and biases)

    Arguments
        nn_architecture: a list of integers
                         length of the list is the number of layers
                         each item in the list specifies
                         the number of units in that layer
    Returns
        weights: list of numpy arrays, w0, w1, ...
        biases: list of numpy arrays, b0, b1, ...

    '''
    weights = []
    biases = []

    # SOLUTION CODE BEGIN #
    for l in range(0, len(nn_architecture)-1):
        w = np.random.randn(nn_architecture[l], nn_architecture[l+1])
        weights.append(w)
        b = np.random.randn(1, nn_architecture[l+1])
        biases.append(b)
    # SOLUTION CODE END #

    return weights, biases

def sigmoid(x):
    '''
    Calculates the sigmoid of an input
    Arguments
        x: np.array() of real numbers
    Returns
        np.array() of real numbers between 0 and 1

    '''
    # Applying sigmoid function element-wise to the input array
    return 1 / (1 + np.exp(-x))

def predict(X, weights, biases):
    '''
    Given an input and NN parameters,
    this function calculates the output

    Arguments
        X: np.array() of shape
        (#datapoints, #features)
        weights: list of np arrays
        biases: list of np arrays
    Returns
        y_pred: np.array() of shape
        (#datapoints, #classes)

    '''

    # SOLUTION CODE BEGIN #
    layer_input = X
    for i in range(len(weights)):
        layer_output = sigmoid(np.dot(layer_input, weights[i]) + biases[i])
        layer_input = layer_output
    y_pred = layer_output
    # SOLUTION CODE END #

    return y_pred

# Task 4
def accuracy(y_pred_binary, y):
    '''
    Given
    -binary predictions
    -labels
    calculates accuracy

    '''

    # SOLUTION CODE BEGIN #
    return sum(y_pred_binary == y)/len(y)
    # SOLUTION CODE END #


def loss_function(X, y, weights, biases):
    losses = []
    for i in range(len(X)):
        prediction = predict(X[i], weights, biases)
        # Cross-entropy loss calculation
        loss = -np.sum(y[i] * np.log(prediction))
        losses.append(loss)

    return np.mean(losses)

# Task 6.1
def weight_update(weights, layer_id, i, j, new_param):
    '''
    Given a single weight defined by (layer_id, i, j)
    updates it with a new parameter (new_param)

    Arguments
        weights: list of np.arrays()
        layer_id: integer
        i: integer
        j: integer
        new_param: float
    Return
        new_w: list of np.arrays()
               where only a single value is changed
               compared to weights

    '''
    # It is important to create a copy of the weights and update the copy
    new_w = deepcopy(weights)

    # SOLUTION CODE BEGIN #
    layer_w = weights[layer_id]
    new_layer_w = layer_w.copy()
    new_layer_w[i, j] = new_param
    new_w[layer_id] = new_layer_w
    # SOLUTION CODE END #

    return new_w

# Task 6.2
def bias_update(biases, layer_id, j, new_param):
    '''
    Given a single bias defined by (layer_id, j)
    updates it with a new parameter (new_param)

    Arguments
        biases: list of np.arrays()
        layer_id: integer
        j: integer
        new_param: float
    Return
        new_b: list of np.arrays()
               where only a single value is changed

    '''

    # It is important to create a copy and update the copy
    new_b = deepcopy(biases)

    # SOLUTION CODE BEGIN #
    layer_b = biases[layer_id]
    new_layer_b = layer_b.copy()
    new_layer_b[0, j] = new_param
    new_b[layer_id] = new_layer_b
    # SOLUTION CODE END #

    return new_b

# Task 6.3
def gradient_estimator(X, y, weights, biases):
    '''
    This function estimates

    ∂(loss_function)/∂(w_ij) and
    ∂(loss_function)/∂(b_ij)

    for every weight and bias.

    ∂(loss_function)/∂(w_ij) can be estimated by
    [(loss for w_ij+epsilon) - (loss with w_ij)]/epsilon

    likewise

    ∂(loss_function)/∂(b_ij) can be estimated by
    [(loss for b_ij+epsilon) - (loss with b_ij)]/epsilon

    For each parameter in weights and biases,
    the function will estimate
    ∂(loss_function)/∂(w_ij) and ∂(loss_function)/∂(b_ij)
    and return a list of np.arrays()
    called w_pds and b_pds, respectively.

    Arguments
        X: input data
        y: labels
        weights: list of np arrays, weights of the MLP
        biases: list of np arrays, biases of the MLP
    Returns
        w_pds: list of np arrays, same dimensions with weights
        b_pds: list of np arrays, same dimensions with biases
    '''
    eps = 1e-4
    loss_val = loss_function(X, y, weights, biases)

    # partial derivatives for weights
    w_pds = []
    for layer_id, layer_w in enumerate(weights):
        pd_array = np.zeros(layer_w.shape)

        # Calculate the partial derivative (pd) for each weight
        # and save those values into pd_array

        # SOLUTION CODE BEGIN #
        for i in range(layer_w.shape[0]):
            for j in range(layer_w.shape[1]):

                wij = layer_w[i, j]

                wij_plus = wij + eps
                w_plus = weight_update(weights, layer_id, i, j, wij_plus)
                loss_plus = loss_function(X, y, w_plus, biases)

                pd = (loss_plus - loss_val)/eps
                pd_array[i, j] = pd
        # SOLUTION CODE END #

        w_pds.append(pd_array)

    # partial derivatives for biases
    b_pds = []
    for layer_id, layer_b in enumerate(biases):
        pd_array = np.zeros(layer_b.shape)

        # Calculate the partial derivative (pd) for each bias
        # and save those values into pd_array

        # SOLUTION CODE BEGIN #
        for j in range(layer_b.shape[1]):
            i = 0
            bij = layer_b[i, j]

            bij_plus = bij + eps
            b_plus = bias_update(biases, layer_id, j, bij_plus)
            loss_plus = loss_function(X, y, weights, b_plus)

            pd = (loss_plus - loss_val)/eps
            pd_array[i, j] = pd
        # SOLUTION CODE END #

        b_pds.append(pd_array)

    return w_pds, b_pds

def one_step_gradient_descent(X, y, weights, biases, lr):
    '''
    This function executes one step of gradient decent and updates weights and biases.

    First estimate the gradient using the gradient_estimator function implemented
    in the previous task.

    Then, update the params by:
    w = w - lr * w_pds
    b = b - lr * b_pds

    Arguments:
        X: input data
        y: labels
        weights: NN weights
        biases: NN biases
        lr: learning rate, a float
    Returns:
        new_w: updated weights, same size as weights
        new_b: updated biases, same size as biases

    '''
    w_pds, b_pds = gradient_estimator(X, y, weights, biases)

    new_w = []
    # For each weight array in weights list
    # use gradient descent update rule
    # append the array to new_w list

    # SOLUTION CODE BEGIN #
    for layer_id, layer_w in enumerate(weights):
        layer_new_w = layer_w - lr*w_pds[layer_id]
        new_w.append(layer_new_w)
    # SOLUTION CODE END #

    new_b = []
    # For each bias array in biases list
    # use gradient descent update rule
    # append the array to new_b list

    # SOLUTION CODE BEGIN #
    for layer_id, layer_b in enumerate(biases):
        layer_new_b = layer_b - lr*b_pds[layer_id]
        new_b.append(layer_new_b)
    # SOLUTION CODE END #

    return new_w, new_b

def gradient_descent(X, y, weights, biases, epoch, lr):
    '''
    Repeats one_step_gd()
    for given number of epochs
    Saves and plots the error at each epoch
    Arguments
        X: input data
        y: labels
        weights: NN weights
        biases: NN biases
        epoch: integer
        lr: float
    Return
        weights: learned weights
        biases: learned biases
    '''

    errors = []

    for i in range(epoch):

        # SOLUTION CODE BEGIN #
        weights, biases = one_step_gradient_descent(X, y, weights, biases, lr)
        error = loss_function(X, y, weights, biases)
        print("Epoch " + str(i) + ": " + str(error))
        # SOLUTION CODE END #

        errors.append(error)

    plt.figure(dpi=200)
    plt.plot(errors)
    plt.xlabel('# epochs')
    plt.ylabel('loss')

    return weights, biases
````

To manage time and computational resources effectively, I train the model for only 5 epochs. While training the model, we are displaying the loss in each epoch and then a graph that shows the learning progress of the model.

````python
# Define MLP architecture
nn_architecture = [X_train_reshaped.shape[1], 64, 8]

# Initialize weights and biases
weights, biases = initializer(nn_architecture)

# Train the MLP using gradient descent
trained_weights, trained_biases = gradient_descent(X_train_reshaped, y_train, weights, biases, epoch=5, lr=0.01)
````
