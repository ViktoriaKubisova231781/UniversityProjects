## WER in Assembly

---

In this section, we will calculate the Word Error Rate (WER) for the Assembly API by loading the data from an Excel file and computing the WER using substitutions, insertions, and deletions relative to the total number of tokens.

```python
# Load the Assembly data
df = pd.read_excel("STT_Assembly.xlsx").dropna()

df
```

```python
### Calculate WER

# Sum the values for Substitution, Insertion and Deletion
S = df["S"].sum()
I = df["I"].sum()
D = df["D"].sum()

# Calculate the total number of tokens (without punctuation)
N = sum(len([w for w in word_tokenize(s) if w not in string.punctuation]) for s in df["Sentence"].values)
N = N + D - I      # As instructed in creative brief: N = (total token number) + deletion - insertion
# Note: Punctuation marks are excluded from WER calculations to focus solely on word-level transcription accuracy, 
# as punctuation is subjective and not directly related to speech recognition performance.

# Calculate WER
WER = (S + I + D) / N

# Report
print("Substitutions = ", S)
print("Insertions = ", I)
print("Deletions = ", D)
print("Total Tokens (Ground Truth) = ", N)
print("WER = ", WER)
```

## WER in Whisper

---

In this section, we will calculate the Word Error Rate (WER) for the Whisper model by loading the data from an Excel file and computing the WER using the same methodology of counting substitutions, insertions, and deletions relative to the total token count.

```python
# Load the Whisper data
df = pd.read_excel("STT_Whisper.xlsx").dropna()

df
```

```python
### Calculate WER

# Sum the values for Substitution, Insertion and Deletion
S = df["S"].sum()
I = df["I"].sum()
D = df["D"].sum()

# Calculate the total number of tokens
N = sum(len([w for w in word_tokenize(s) if w not in string.punctuation]) for s in df["Sentence"].values)
N = N + D - I      # As instructed in creative brief: N = (total token number) + deletion - insertion
# Note: Punctuation marks are excluded from WER calculations to focus solely on word-level transcription accuracy, 
# as punctuation is subjective and not directly related to speech recognition performance.

# Calculate WER
WER = (S + I + D) / N

# Report
print("Substitutions = ", S)
print("Insertions = ", I)
print("Deletions = ", D)
print("Total Tokens (Ground Truth) = ", N)
print("WER = ", WER)
```

### Detailed Error Analysis

The local Whisper model demonstrated marginally better performance with a WER of 2.64%, outperforming the Assembly AI API by 0.50 percentage points. Breaking down the errors:

__Assembly AI:__
- **Substitutions (13)**: Second most common error type
- **Deletions (16)**: Highest among both models
- **Insertions (11)**: Similar to Whisper
- **Total Tokens (Ground Truth)**: 1,273 processed

__Whisper:__
- **Substitutions (11)**: Slightly better than Assembly
- **Deletions (12)**: 25% fewer than Assembly
- **Insertions (12)**: Comparable to Assembly
- **Total Tokens (Ground Truth)**: 1,328 processed

Overall, Whisper performed better, making fewer errors overall in substitutions and deletions, resulting in a lower WER score.

<br>

### Transcription Style Differences

An interesting observation emerges in how each model handles sentence segmentation:

- **Whisper**: Tends to combine related phrases into longer sentences, resulting in:
  - More natural-flowing text
  - Fewer but longer segments
  - Higher total token count (1,328 tokens)

- **Assembly AI**: Favors shorter, more segmented transcriptions, leading to:
  - More distinct sentence breaks
  - More numerous but shorter segments
  - Lower total token count (1,273 tokens)

This difference in segmentation approach explains the variation in total token count while maintaining similar accuracy levels.
