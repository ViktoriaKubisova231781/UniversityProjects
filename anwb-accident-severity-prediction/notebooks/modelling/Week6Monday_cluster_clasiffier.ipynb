# Clustering junctions 
Looking at the map above it seems like the dataset describes "road ends" instead off junctions; since we are only interested in the junctions i am going to use DB scan to group these "junction clusters" togheter 

```python
from sklearn.cluster import DBSCAN

# Create a numpy array of junction coordinates
junction_coords = junctions_df[['Longitude', 'Latitude']].values

# Perform DBSCAN clustering
clustering = DBSCAN(eps=0.0003, min_samples=1).fit(junction_coords)

# Get the cluster labels
cluster_labels = clustering.labels_

# Add the cluster labels to the junctions_df
junctions_df['junction_id'] = cluster_labels

# Now, you can use the junctions_df to display one location per Junction_id 
junctions_df
```

# Labeling the accidents

```python
import geopandas as gpd
import pandas as pd
from scipy.spatial import cKDTree

# Assuming your DataFrames are named 'accidents_df' and 'junctions_df'
# Assuming your accident coordinates are in 'X' and 'Y' columns

# Calculate average latitude and longitude for each junction cluster (assuming 'cluster_id' exists)
junctions_df['center_lat'] = junctions_df.groupby('junction_id')['Latitude'].transform('mean')
junctions_df['center_lon'] = junctions_df.groupby('junction_id')['Longitude'].transform('mean')

# Optional: Drop duplicates if 'junction_id' uniquely identifies junctions
# If 'junction_id' might have duplicates within clusters, keep all rows in 'junctions_df'
single_junctions_df = junctions_df.drop_duplicates(subset='junction_id') 

# Create GeoDataFrames from your DataFrames
accidents_gdf = gpd.GeoDataFrame(
    accidents_df, 
    geometry=gpd.points_from_xy(accidents_df['Longitude'], accidents_df['Latitude']), 
    crs="EPSG:3857"
)
single_junctions_gdf = gpd.GeoDataFrame(
    single_junctions_df,  # Pass the actual data
    geometry=gpd.points_from_xy(single_junctions_df['center_lon'], single_junctions_df['center_lat']), 
    crs="EPSG:3857"
) 

# perform nearest-neighbor join
nearest_junctions = gpd.sjoin_nearest(accidents_gdf, single_junctions_gdf, max_distance=60)

accidents_df['junction_id'] = nearest_junctions['junction_id']

# Explore the result (optional)
print(accidents_df.head())
```

# Attribute based clustering

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import folium

# Ensure necessary libraries are imported
import geopandas as gpd
from scipy.spatial import cKDTree

# Columns to use for clustering
columns_to_cluster = ['Accident_Severity', 'Parties_Involved', 'Year']

# Select the relevant columns for clustering
data_to_cluster = df_cleaned[columns_to_cluster]

# Handle missing values
data_to_cluster = data_to_cluster.dropna()

# Verify the data to be clustered is not empty
print("Shape of the data to be clustered:", data_to_cluster.shape)
print(data_to_cluster.head())

# Standardize the data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_to_cluster)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=5, random_state=42)  # Adjust the number of clusters as needed
data_to_cluster['Cluster'] = kmeans.fit_predict(data_scaled)
df_cleaned['Attribute_Cluster'] = data_to_cluster['Cluster']

# Analyze the clusters
print(data_to_cluster['Cluster'].value_counts())

# Visualize clusters using pairplot
sns.pairplot(data_to_cluster, hue='Cluster', palette='viridis')
plt.show()

# Example of visualizing cluster centers
cluster_centers = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=data_to_cluster.columns[:-1])
print(cluster_centers)

# Summarize clusters
cluster_summary = data_to_cluster.groupby('Cluster').mean()
print(cluster_summary)

# Example of cross-tabulation
cross_tab = pd.crosstab(df_cleaned['Attribute_Cluster'], df_cleaned['Accident_Severity'])
print(cross_tab)
```

# Visualisation of the clusters:

```python
# Function to plot a specific cluster
def plot_cluster(cluster_id):
    cluster_data = df_cleaned[df_cleaned['Attribute_Cluster'] == cluster_id]
    
    m = folium.Map(location=[cluster_data['Latitude'].mean(), cluster_data['Longitude'].mean()], zoom_start=14)
    colors = ['blue', 'green', 'red', 'purple', 'orange']
    
    for index, row in cluster_data.iterrows():
        folium.CircleMarker(location=[row['Latitude'], row['Longitude']], 
                            radius=4, 
                            color=colors[cluster_id], 
                            fill=True, 
                            fill_color=colors[cluster_id]).add_to(m)
    return m

# Plot each cluster separately
for cluster_id in range(5):
    m = plot_cluster(cluster_id)
    display(m)  # This function works in Jupyter Notebook to display the map
```

# Analysis of the clusters:

```python
# Function to analyze a specific cluster
def analyze_cluster(cluster_id):
    cluster_data = df_cleaned[df_cleaned['Attribute_Cluster'] == cluster_id]
    
    # Summary statistics
    summary_stats = cluster_data.describe()
    print(f"Summary statistics for cluster {cluster_id}:")
    print(summary_stats)
    
    # Visualize distributions
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Attribute_Cluster', y='Accident_Severity', data=cluster_data)
    plt.title(f'Distribution of Accident Severity in Cluster {cluster_id}')
    plt.show()
    
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Attribute_Cluster', y='Parties_Involved', data=cluster_data)
    plt.title(f'Number of Parties Involved in Cluster {cluster_id}')
    plt.show()
    
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Attribute_Cluster', y='Year', data=cluster_data)
    plt.title(f'Year Distribution in Cluster {cluster_id}')
    plt.show()

# Analyze each cluster
for cluster_id in range(5):
    analyze_cluster(cluster_id)
```

# Developing a Classification Model:

## Preparing the data

```python
from sklearn.model_selection import train_test_split

# Select features and target
features = df_cleaned[['Accident_Severity', 'Parties_Involved', 'Year']]
target = df_cleaned['Attribute_Cluster']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

## Training the model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Initialize the classifier
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
clf.fit(X_train, y_train)

# Make predictions
y_pred = clf.predict(X_test)
```

## Model Evaluation:

```python
# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
